```python
from imageai.Detection.Custom import DetectionModelTrainer
from google.colab import drive
```

```python
trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
```

```python
trainer.setDataDirectory(data_directory="/content/drive/MyDrive/ai_dataset")
trainer.setTrainConfig(object_names_array=[], batch_size=4, num_experiments=25, train_from_pretrained_model="pretrained-yolov3.h5")
trainer.trainModel()
```

```python
trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/content/drive/MyDrive/ai_dataset")
trainer.evaluateModel(model_path="/content/drive/MyDrive/ai_dataset/models", json_path="/content/drive/MyDrive/ai_dataset/json/detection_config.json", iou_threshold=0.5, object_threshold=0.3, nms_threshold=0.5)
```
